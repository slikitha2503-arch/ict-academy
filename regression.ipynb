```python
#regression example: predict house price based on area
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
#home rental
df=pd.read_csv('/content/home-rental.csv',encoding='latin1')
print(df.head())
```

```python
#features and target
x=df[['postal_code']]      #independebt variable
y=df['rent_amount']    #dependent variable
```

```python
#split into training and testing
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
#create a regression  model
model=LinearRegression()
model.fit(x_train,y_train)
```

```python
#predict
y_pred=model.predict(x_test)
```

```python
#evaluate
mse=mean_squared_error(y_test,y_pred)
print('mean squared error:',mse)
```

```python
plt.scatter(x,y,color='blue')
plt.plot(x, model.predict(x),color='red')
plt.xlabel('postal_code')
plt.ylabel('rent_amount')
plt.title('Linear regression')
plt.show()
```

```python
#agglomerative clustering example with csv
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import AgglomerativeClustering
```

```python
# 1. load csv data
#replace 'your_file.csv' with the path to your csv file
df=pd.read_csv('/content/Mall_Customers.csv')
```

```python
# 2. select numerical columns for clustering
#replac these colums with numerical column in your csv
x=df[['Annual Income (k$)','Spending Score (1-100)']]
```

```python
#3 create agglomerative clustering model
agg_model=AgglomerativeClustering(n_clusters=3)
```

```python
#4 fit model and assign cluster labels
df['cluster']=agg_model.fit_predict(x)
```

```python
#5 plot cluster
plt.scatter(x['Annual Income (k$)'],x['Spending Score (1-100)'],c=df['cluster'],cmap='rainbow')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.title('Agglomerative Clustering')
plt.show()
```

```python
#6 optional: save clustered dataset
df.to_csv('clustered_data.csv',index=False)
print("Agglomerative clustered dataset saved as agglo_cluatered_data.csv")
```
