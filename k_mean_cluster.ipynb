```python
#k-mean clustering example: group customer
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
#sample data
df=pd.read_excel('/content/sample_data/Mall_Customers.xlsx')
print(df.head())
```

```python
#create k-mean model
kmeans=KMeans(n_clusters=3, random_state=42)
df['cluster']=kmeans.fit_predict(df[['Annual Income', 'Spending Score']])
```

```python
#centroids
centroids=kmeans.cluster_centers_
```

```python
#plot clusters
plt.scatter(df['Annual Income'], df['Spending Score'], c=df['cluster'],cmap='rainbow')
plt.scatter(centroids[:,0], centroids[:,1], color='black', marker='x',s=100)
plt.xlabel('Annual Income')
plt.ylabel('spending score')
plt.title('k-mean clusterng example')
plt.show()
```
